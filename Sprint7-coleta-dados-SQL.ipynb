```python
from scipy import stats

df_07 = pd.read_csv('/datasets/project_sql_result_07.csv')

display(df_07.head())
df_07.info()
```

```python
good = df_07[df_07['weather_conditions'] == 'Good']['duration_seconds']
bad  = df_07[df_07['weather_conditions'] == 'Bad']['duration_seconds']

len(good), len(bad), good.mean(), bad.mean()
```

### Teste de hipótese: impacto do clima na duração das corridas

- **H0 (hipótese nula):** a duração média das viagens Loop → O'Hare é a mesma em sábados com clima **Good** e **Bad**.  
- **H1 (hipótese alternativa):** a duração média das viagens Loop → O'Hare é **diferente** em sábados com clima **Good** e **Bad**.

Escolhi **α = 0.05**, um valor padrão que controla o risco de falso positivo em 5%.


```python
alpha = 0.05
stat, p_value = stats.ttest_ind(good, bad, equal_var=False)

stat, p_value
```

```python
if p_value < alpha:
    print("Rejeitamos H0: há evidência de diferença na duração média entre Good e Bad.")
else:
    print("Não rejeitamos H0: não há evidência estatística suficiente de diferença na duração média.")
```

**Resultado:**  
O teste t de Welch retornou um *p-value* extremamente pequeno (p < 0.001), menor que o nível de significância α = 0.05.  
Portanto, rejeitamos a hipótese nula (H0) e concluímos que **a duração média das corridas do Loop até o Aeroporto Internacional O'Hare é significativamente diferente em sábados com clima Good e Bad**.

Isso indica que as condições climáticas têm impacto estatisticamente significativo na duração das viagens.


Na prática, os resultados mostram que o clima tem impacto real na duração das corridas.  
Em sábados com clima ruim (chuva ou tempestade), as viagens do Loop até o Aeroporto Internacional O’Hare tendem a durar mais tempo do que em condições climáticas favoráveis.

Isso pode ser explicado por fatores como trânsito mais intenso e redução da velocidade média em dias de chuva.  
Portanto, o clima é um fator relevante que deve ser considerado no planejamento operacional e na estimativa do tempo de viagem.
